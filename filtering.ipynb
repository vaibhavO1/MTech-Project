```python
import numpy as np
import scipy.io.wavfile as wav
from scipy.signal import butter, filtfilt, resample
import matplotlib.pyplot as plt
```

```python
# Schmidt's algorithm (pseudo code implementation)
# This is a simplified version; the actual implementation would depend on specific requirements.
def schmidt_spike_removal(signal, threshold=0.05):
    cleaned_signal = signal.copy()
    for i in range(1, len(signal)-1):
        if np.abs(signal[i] - signal[i-1]) > threshold and np.abs(signal[i] - signal[i+1]) > threshold:
            cleaned_signal[i] = (signal[i-1] + signal[i+1]) / 2  # Replace spike with the average of neighbors
    return cleaned_signal
```

```python
# Load the PCG signal
sample_rate, signal = wav.read('/content/normal__201102201230.wav')
```

```python
# Step 1: Butterworth bandpass filtering (20 Hz - 900 Hz)
def butter_bandpass_filter(data, lowcut, highcut, fs, order=6):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    y = filtfilt(b, a, data)
    return y
```

```python
filtered_signal = butter_bandpass_filter(signal, 20, 900, sample_rate)
```

```python
# Step 2: Downsampling to 2000 Hz
target_sample_rate = 2000
if sample_rate != target_sample_rate:
    downsampled_signal = resample(filtered_signal, int(len(filtered_signal) * target_sample_rate / sample_rate))
    sample_rate = target_sample_rate
else:
    downsampled_signal = filtered_signal
```

```python
# Step 3: Friction spike removal using Schmidt's algorithm
cleaned_signal = schmidt_spike_removal(downsampled_signal)
```

```python
# Step 4: Segmentation using a 2-second window with 75% overlap
window_size = 2 * sample_rate  # 2 seconds window
step_size = int(window_size * 0.25)  # 75% overlap
```

```python
# Plot the segmented results
plt.figure(figsize=(10, 8))
plt.subplot(3, 1, 1)
plt.title("Original PCG Signal")
plt.plot(signal)
plt.subplot(3, 1, 2)
plt.title("Filtered PCG Signal")
plt.plot(filtered_signal)
plt.subplot(3, 1, 3)
plt.title("Downsampled and Cleaned PCG Signal")
plt.plot(cleaned_signal)
plt.show()
```
